```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import squarify
```

```python
df = pd.read_csv(r"C:\\Users\\ritad\Documents\\PSVDados\\psvdata.csv")
df.head()
```

```python
dataGA = df[df["GA"]>2]
```

```python
norm = matplotlib.colors.Normalize(vmin=min(dataGA.GA), vmax=max(dataGA.GA))
```

```python
colors = [matplotlib.cm.Reds(norm(value)) for value in dataGA.GA]
```

```python
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 4.5)

squarify.plot(label = dataGA.Player, sizes = dataGA.GA, color = colors, alpha=.9)
plt.title("PSV: Participações diretas em gols em 2021-22 (todas as competições)", fontsize=20,fontweight="bold")

plt.axis('off')
plt.show()
```
